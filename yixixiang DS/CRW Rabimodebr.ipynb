```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from qutip import*
```

```python
wc = 1.0*2.0*np.pi
wa = 1.0*2.0*np.pi
g = 0.25*2*np.pi
kappa = 0.005
gamma = 0.05
N = 50
n_th_a=0.0
tlist = np.linspace(0.0, 1000,6000)
```

```python
psi0 = tensor (basis(N,0), basis(2))
a = tensor(destroy(N), qeye(2))
sm = tensor(qeye(N), destroy(2))
H = wc*a.dag()*a+wa*sm.dag()*sm+g*(a.dag()+a)*(sm+sm.dag())
H
```

```python
from numpy import*
def spectrum_cavity(w):
    if w > 0: 
        return kappa 
    else:
        return 0.0 

def spectrum_atom(w):
    if w > 0:
        return gamma
    else:
        return 0.0
a_ops = [
    [a + a.dag(), spectrum_cavity], 
    [sm + sm.dag(), spectrum_atom]
    ]
```

```python
output = brmesolve(H,psi0,tlist,a_ops,e_ops=[a.dag()*a,sm.dag()*sm])
output
```

```python
n_c = output.expect[0]
n_a = output.expect[1]
fig,axes = plt.subplots(1,1,figsize=(10,5))
axes.plot(tlist,n_c,label='Cavity')
axes.plot(tlist,n_a,label='Atom excited state')
axes.legend()
axes.set_xlabel('Time [s]')
axes.set_ylabel('Occupation probability')
axes.set_title('CRW Vacuum Rabi oscillations')
```
